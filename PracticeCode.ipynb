* How would you store the year (1961) and name ("Gagarin") together so they cannot be changed later?

* Can you change the year in moon_info = (1969, "Armstrong") to 1970 after you create it? (Yes or No)

```python
astronauts = (1961, "Gagarin")
moon_info = (1969, "Armstrong")
print(moon_info[0])
moon_info = (1970, "Armstrong")
print(moon_info)
#It got changed? Wut.
```

You're right to notice that `moon_info` seems to change, but it's crucial to understand the difference between **modifying a tuple** and **reassigning a variable**.

1.  **Tuples are Immutable**: The original tuple `(1969, "Armstrong")` *itself* cannot be changed. You cannot, for example, do `moon_info[0] = 1970` – that would raise an error.
2.  **Variable Reassignment**: What happens with `moon_info = (1970, "Armstrong")` is that you are creating a *brand new tuple* `(1970, "Armstrong")` and then making the variable `moon_info` *point to this new tuple*. The original tuple `(1969, "Armstrong")` still exists in memory (for a short while), but the variable `moon_info` no longer refers to it. It now refers to the new tuple. If nothing else refers to the original tuple, it will eventually be removed by Python's garbage collector.

So, you didn't *change* the existing tuple; you replaced the variable's reference with a *new* tuple.

```python
# Create a function which can identify if the triangle is equilateral, acute, and obtuse.
# For the eq., all three sides are equal. For the acute one, all three angles are less than 90 degrees. For the obtuse, all angles are more than 90 degrees.

def triangle(a, b, c):
  if a+b+c!=180 or a<=0 or b<=0 or c<=0:
    return ("Invalid!")

  if a==60 and b==60 and a==60:
    print(f"The triangle is equilateral.")
  elif max(a, b, c)<90:
    print(f"The triangle is acute.")
  elif max(a, b, c)==90:
    print(f"It is a right-angled triangle.")
  else:
    print(f"The triangle is obtuse.")
```

```python
triangle(60, 60, 60)
```

* Magnitude Difference: Create a function to calculate the magnitude difference between two stars (magnitude1 - magnitude2).
* Distance Modulus: Create a function to calculate the distance modulus (apparent magnitude - absolute magnitude).

* Sphere Area: Create a function to calculate the surface area of a sphere given its radius.
* Leap Year: Create a function that determines whether a given year is a leap year.

```python
def mag_diff(star01, star02):
  return star01-star02

star01 = float(input("What is the distance for the first star? "))
star02 = float(input("What is the distance for the second star? "))
print(f"The difference between the two is {round(mag_diff(star01, star02), 4)} AU.")
```

```python
def dist_mod(app, ab):
  return app-ab

app = float(input("What is the apparent magnitude? "))
ab = float(input("What is the absolute magnitude "))
print(f"The distance modulus is {round(dist_mod(app, ab), 4)}.")
```

```python
w_constant = 2.898e-3
def weins(temp):
  return w_constant/temp

temp = float(input("What is the temperature of the radiation? (In Kelvin) "))
print(f"The peak wavelength of the radiation is {weins(temp)} m.")
```

```python
def surface_area(radius):
  return 4*np.pi*pow(radius, 2)

radius = float(input("What is the radius of the sphere? "))
print(f"The surface area of the sphere is {round(surface_area(radius), 3)} square meters.")
```

```python
def leap(year):
  if year%4==0 and year%100==0:
    return "It is a leap year!"
  else:
    return "Not a leap year!"

year = int(input("What is the year in question? "))
leap(year)
```
